```python
import matplotlib.pyplot as plt
from math import pi, cos, sin
```

```python
CNT = 1024
X = [2 * x * pi / CNT - pi for x in range(CNT)]
C = [cos(x) for x in X]
S = [sin(x) for x in X]
```

```python
# figsize (breite, hoehe)
plt.figure(figsize=(10, 6))

plt.xlim(1.1*min(X), 1.1*max(X))
plt.xlim(1.1*min(X), 1.1*max(X))

plt.xticks([-pi, -pi / 2 , 0, pi/2, pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
plt.yticks([-1, -0.5, 0, 0.5, 1], [r'$-1$', r'$-0.5$', r'$0$', r'$0.5$', r'$+1$'])

# achsen
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

t = 2*pi/3
plt.plot([t,t],[0, sin(t)], color ='red', linewidth=2.5, linestyle="--")
plt.scatter([t],[sin(t)], 50, color ='red')

plt.plot([t,t],[0, cos(t)], color ='blue', linewidth=2.5, linestyle="--")
plt.scatter([t],[cos(t)], 50, color ='blue')

plt.annotate(r'$\sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
xy=(t, sin(t)), xycoords='data',
xytext=(20, 10), textcoords='offset points', fontsize=16,
arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.annotate(r'$\cos(\frac{2\pi}{3})=-\frac{1}{2}$',
xy=(t, cos(t)), xycoords='data',
xytext=(20, 10), textcoords='offset points', fontsize=16,
arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.plot(X, C, color='blue', linestyle=':', linewidth=2.5, label="cos")
plt.plot(X, S, color='red', linestyle='-.', linewidth=2.5, label='sin')
plt.legend(loc='upper left', frameon=False)
plt.title('Niks Plot')
```
